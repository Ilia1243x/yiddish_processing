```python
import urllib.request
from bs4 import BeautifulSoup
import requests
from urllib.request import Request, urlopen
import tqdm
import glob
import pandas as pd
from IPython.display import clear_output
import os
import sys
```

```python
import cloudscraper
import time
import random
```

```python
scraper = cloudscraper.create_scraper()

url1 = "https://www.yiddishbookcenter.org/search?f%5B0%5D=collection%3AYiddish%20Book%20Center%27s%20Spielberg%20Digital%20Yiddish%20Library&page=0"
base_url = "https://www.yiddishbookcenter.org"
books = []

for page in range (365, 1000):
  url = f"{url1}{page}"
  response = scraper.get(url)
  #time.sleep(1)
  soup = BeautifulSoup(response.text, 'html.parser')
  Z = soup.findAll('a', class_ = "search-result search-result--yiddish")
  for data in Z:
    title = data.find('h3', class_ = "search-result__title-trans" ).text
    author = data.find('p', class_ = "search-result__creator-authority" ).text
    next_link = data.get('href')
    urllink = base_url+next_link
    book_page = scraper.get(urllink)
    #time.sleep(random.uniform(2,3))
    saulgoodman = BeautifulSoup(book_page.text, 'html.parser')
    arc_link = saulgoodman.find('a', class_ = "icon-link icon-link--reverse")
    final_link = arc_link.get('href')
    books.append({
              "title": title,
              "author": author,
              "link": final_link
    })
df = pd.DataFrame(books)
df
```

```python
df.to_csv("books_info.csv", index=False, encoding='utf-8-sig')
```

```python
scraper = cloudscraper.create_scraper()
myset = dict()
for name, link in zip(df['title'][5100:], df['link'][5100:]):
  myset[name] = myset.get(name, 0) + 1
  pdf_response = scraper.get(link)
  if (pdf_response.status_code != 200):
    print("Skipped")
    continue
  title = f"{name}{myset[name]}.pdf".replace("/","_")
  title = title.replace(" ", "_")
  time.sleep(random.uniform(1,2))
  path = os.path.join(save_dir, title)
  with open(path, 'wb') as pdf_file:
    pdf_file.write(pdf_response.content)
```
